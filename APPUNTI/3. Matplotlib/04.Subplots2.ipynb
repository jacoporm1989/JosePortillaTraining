```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Riprendiamo in nostro subplots avente 3 righe e 3 colonne:
```

```python
fig,axes= plt.subplots(3,3) 
```

```python
# Si noti che i numeri sono scritti malissimo: infatti, molti di loro sono 
# sovrapposti. Per migliorare il layout, si utilizza plt.tight_layout():
```

```python
fig,axes= plt.subplots(3,3) 
plt.tight_layout()
```

```python
# MOLTO MEGLIO!!!
```

```python
# Possiamo anche aumentare lo spazio tra un quadrato e l'altro in senso orizzontale
# usando la funzione fig.subplots_adjust(wspace=a), con a compreso
# tra 0 e 1. ESEMPIO
```

```python
fig,axes= plt.subplots(3,3) 
plt.tight_layout()
fig.subplots_adjust(wspace=0.4)
```

```python
fig,axes= plt.subplots(3,3) 
plt.tight_layout()
fig.subplots_adjust(wspace=0.6)
```

```python
fig,axes= plt.subplots(3,3) 
plt.tight_layout()
fig.subplots_adjust(wspace=0.9)
```

```python
# Ovviamente, mano mano che aumenta lo spazio tra un quadrato e l'altro, si riduce
# la larghezza dell'asse delle x.
# Ragioniamo ora in verticale. Se vogliamo aumentare lo spazio tra un quadrato e l'altro
# in senso verticale dovremo usare fig.subplots_adjust(hspace=a), 0<a<1
```

```python
fig,axes= plt.subplots(3,3) 
plt.tight_layout()
```

```python
fig,axes= plt.subplots(3,3) 
plt.tight_layout()
fig.subplots_adjust(hspace=0.4)
```

```python
fig,axes= plt.subplots(3,3) 
plt.tight_layout()
fig.subplots_adjust(hspace=0.8)
```

```python
# Mano mano che aumenta la distanza tra un quadrato e l'altro in senso verticale,
# si avr una diminuzione della lunghezza degli assi delle y.
```

```python
# wspace e hspace possono anche essere inseriti insieme:
```

```python
fig,axes= plt.subplots(3,3) 
fig.subplots_adjust(wspace=0.3,hspace=0.5)
plt.tight_layout()
```

```python
# Vediamo un banale esempio lineare per ripasso
```

```python
np.random.seed(12)
x = np.random.randint(3,10,4)
x
```

```python
y = 3*x
y
```

```python
fig,axes= plt.subplots(3,3) 
plt.tight_layout()
fig.subplots_adjust(wspace=0.7,hspace=0.7)
axes[0][1].plot(x,y)
axes[2][0].plot(x,y)
axes[1][1].plot(x,y)
```

```python
# Inseriamo ora il titolo sull'asse delle x e sull'asse delle y
# dei nostri grafici. Per una migliore visualizzazione, consideriamo
# solo 4 grafici
```

```python
fig,axes= plt.subplots(2,2) 
plt.tight_layout()
fig.subplots_adjust(wspace=0.7, hspace=0.5)
# INSERIMENTO TITOLO GRAFICI


axes[0][0].plot(x,y)

#INSERIMENTO TITOLO SULL'ASSE DELLE X DEL GRAFICO IN POSIZIONE 0,0

axes[0][0].set_xlabel("TRIMESTRE")

#INSERIMENTO TITOLO SULL'ASSE DELLE Y DEL GRAFICO IN POSIZIONE 0,0

axes[0][0].set_ylabel("REDDITO")

# INSERIMENTO TITOLO DEL GRAFICO IN POSIZIONE 0,0

axes[0][0].set_title("Variazione reddito per trimestre")

# INSERIMENTO TITOLO GENERALE PER TUTTI E 4 I GRAFICI

# SE SCRIVI fig.suptitle("ESEMPI"), IL GRAFICO ESCE UNA PORCHERIA. BISOGNA USARE DUE PARAMETRI

fig.suptitle("ESEMPI")
plt.subplots_adjust(top=0.85)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# STUDIAMO ORA L'APPLICAZIONE DEGLI STILI.
# PARTIAMO COME AL SOLITO DALL'EQUAZIONE DELLA RETTA E DAL NOSTRO INTERO CANVAS
```

```python
np.random.seed(10)
x = np.random.randint(0,20,4)
x
```

```python
y = 5*x
y
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])   # Gli assi prendono il 100% del canvas
axes.plot(x,y)
```

```python
# PER CAMBIARE COLORE SI INSERISCE IL PARAMETRO color ALL'INTERNO DELLA FUNZIONE plot
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])   # Gli assi prendono il 100% del canvas
axes.plot(x,y,color='red')  # Possiamo anche scrivere il colore in RGB o ESADECIMALE
```

```python
# Creiamo un'altra funzione, di tipo "rettale" per comodit√†
```

```python
np.random.seed(11)
a = np.random.randint(0,20,4)
a
```

```python
b = 2*a
b
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])   # Gli assi prendono il 100% del canvas
axes.plot(a,b,color='green')  
```

```python
# METTENDOLE INSIEME:
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])   
axes.plot(a,b,color='green', label="Reddito") 
axes.plot(x,y,color='blue',label="Et√†")
axes.legend()
```

```python
# Possiamo anche stilizzare lo spessore di una retta tramite il parametro lw di plot
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])   
axes.plot(a,b,color='green', label="Reddito",lw = 5)   # Scrivere lw oppure linewidth √® la stessa cosa.
axes.plot(x,y,color='blue',label="Et√†",lw = 8)
axes.legend()
```

```python
# Vediamo ora il parametro ls, oppure linestyle (√® la stessa cosa). Questo parametro ci dice come visualizzare le nostre rette
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])   
axes.plot(a,b,color='green', label="Reddito",lw = 5,ls=":")   # Scrivere lw oppure linewidth √® la stessa cosa.
axes.plot(x,y,color='blue',label="Et√†",lw = 8,ls="--")
axes.legend()
```

```python
# Possiamo rappresentare questi punti sul grafico tramite un parametro chiamato marker:
# Per una questione di migliore leggibilit√† del grafico eliminiamo il parametro lw
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])   
axes.plot(a,b,color='green', label="Reddito",marker="^")
```

```python
# La lista completa dei markers sta nella documentazione python.
```

```python
# Per curiosit√† riportiamo anche la seconda retta nel nostro grafico con un marker diverso
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])   
axes.plot(a,b,color='green', label="Reddito", marker="+")  
axes.plot(x,y,color='blue',label="Et√†",marker="^")
```

```python
# Possiamo aumentare la grandezza del marker tramite il parametro markersize: possiamo anche scrivere ms al posto di markersize: √® la stessa cosa
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])   
axes.plot(a,b,color='green', label="Reddito", marker="+",markersize=20)  
axes.plot(x,y,color='blue',label="Et√†",marker="^",ms=20) 
```

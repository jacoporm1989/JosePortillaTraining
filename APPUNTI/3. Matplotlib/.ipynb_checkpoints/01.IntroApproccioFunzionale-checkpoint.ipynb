```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# MATPLOTLIB
```

```python
# Matplotlib è basato su Matlab. Ci sono due 
# approcci diversi per creare un plot
# 1. Metodo funzionale
# 2. Metodo basato sulla OOP
```

```python
# 1. Metodo funzionale
```

```python
# Disegnamo una retta tramite plt.plot. Innanzitutto definiamo una retta
```

```python
np.random.seed(12)
x = np.random.randint(10,20,5)
x
```

```python
# l'equazione della retta è y=mx+q. Ponendo 
# m = 3 e q = 10:
```

```python
y = 3*x + 10
y
```

```python
# Rappresentiamolo!
```

```python
plt.plot(x,y)
```

```python
# matplotlib.lines.Line2D è il tipo dell'oggetto e 0x23f1972c590 è lo spazio
#di memoria occupato. Ci sono due modi per non visualizzare questi elementi
```

```python
# Metodo 1
```

```python
plt.plot(x,y);
```

```python
# Metodo 2
```

```python
plt.plot(x,y)
plt.show()
```

```python
# Inserimento titolo del grafico
```

```python
plt.plot(x,y)
plt.title("BLEEEEEE")
```

```python
# Inserimento nomi variabili sull'asse delle x e sull'asse delle y
```

```python
plt.plot(x,y)
plt.title("BLEEEEEE")
plt.xlabel("BLA")
plt.ylabel("BLI")
```

```python
# Si noti che il grafico ci mostra tutta la retta. Possiamo far si che se ne veda
# solo una porzione. Ad esempio, supponiamo di volerla vedere dal punto m (3) al punto 16, sull'asse
# delle x
```

```python
plt.plot(x,y)
plt.title("BLEEEEEE")
plt.xlabel("BLA")
plt.ylabel("BLI")
plt.xlim(13,16)
```

```python
# E ora aggiungiamo i limiti sull'asse delle y:
```

```python
plt.plot(x,y)
plt.title("BLEEEEEE")
plt.xlabel("BLA")
plt.ylabel("BLI")
plt.xlim(3,16)
plt.ylim(54,58)
```

```python
# Non esiste nessuna porzione della retta che segua i criteri stabiliti.
```

```python
# Torniamo alla situazione di partenza
```

```python
plt.plot(x,y)
plt.title("BLEEEEEE")
plt.xlabel("BLA")
plt.ylabel("BLI")
```

```python
# Per salvare il grafico
```

```python
plt.plot(x,y)
plt.title("BLEEEEEE")
plt.xlabel("BLA")
plt.ylabel("BLI")
plt.savefig("primoGrafico.jpg")
```

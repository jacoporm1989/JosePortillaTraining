```python
# Vediamo ora come inscatolare i grafici. Possiamo creare un grafico dentro
# un grafico, nel seguente modo
```

```python
fig = plt.figure()
# GRAFICO ESTERNO
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y)

# GRAFICO INTERNO
axes2 = fig.add_axes([0.3,0.7,0.2,0.2])
axes2.plot(x,y)
```

```python
# La nostra curva verrà rappresentata all'interno del nostro grafico .
# Il punto di partenza sarà il 30% del canvas (e quindi del grafico contenuto) lungo l'asse
# delle x e il 70% lungo l'asse delle y.
# Da quel punto in poi, partiranno l'asse delle x e l'asse delle y del canvas innestato.
# La lunghezza sarà pari al 20% della lunghezza del canvas contenitore ==> è il
# 20% della lunghezza di tutto il canvas.
# Lo stesso ragionamento vale per l'asse delle y.
```

```python
# Nella realtà, il grafico innestato serve per fare uno zoom di una porzione
# della curva rappresentata. Quindi:
```

```python
fig = plt.figure()
# GRAFICO ESTERNO
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y)

# GRAFICO INTERNO
axes2 = fig.add_axes([0.3,0.7,0.2,0.2])
axes2.plot(x,y)
plt.xlim(8,10)
plt.ylim(100000,150000)
```

```python
# FINORA ABBIAMO STUDIATO IL CANVAS E GLI ASSI, MA NON ABBIAMO
# VISTO I PARAMETRI CHE POSSIAMO PASSARE ALLE NOSTRE FUNZIONI.
# TORNIAMO ALLA NOSTRA SITUAZIONE DI PARTENZA.
```

```python
# Partiamo da fig = plt.figure(). I due parametri che vedremo
# sono:
# 1. DPI ==> Mostra quanti inches (pollici) deve essere grande un grafico. 
# Si ricorda che 1 inch = 96px
# 2. FIGSIZE ==> Setta la lunghezza e la larghezza del nostro grafico in pollici.
```

```python
fig = plt.figure(figsize = (4,3), dpi=350)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y)
fig.savefig("graficoinches.jpg")
```

```python
# NB. Il nostro attributo figsize, ossia figsize = (x1,x2) ci dice
# che, se x1>x2 il grafico sarà più largo che alto. Viceversa, se x1<x2, il grafico
# sarò più alto che largo. Se x1=x2 ==> Il grafico sarà quadrato.
```

```python
# Se apriamo il file salvato, quindi nel nostro caso graficoinches.jpg, noteremo
# che i numeri sugli assi delle x e quelli sugli assi delle y sono scomparsi.
# Per farli apparire, aggiungiamo il parametro bbox_inches nella funzione
# fig.savefig() e la valorizziamo con "tight"
```

```python
fig = plt.figure(figsize = (4,3), dpi=350)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y)
fig.savefig("graficoinches2.jpg", bbox_inches="tight")
```

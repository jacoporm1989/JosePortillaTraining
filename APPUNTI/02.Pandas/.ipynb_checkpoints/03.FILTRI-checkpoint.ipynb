```python
import numpy as np
import pandas as pd
```

```python
# I dataframe, nella realtà, sono molto grandi. Questo significa che non si filtra in base
# alla posizione, ma si filtra in base ad una condizione. I filtri ci permettono di selezionare
# le righe in base a delle condizioni applicate alle colonne
```

```python
# Vediamo questi filtri nel dettaglio, considerando il nostro solito Df tips
```

```python
tips = pd.read_csv("tips.csv")
tips.head()
```

```python
# VOGLIAMO ESTRARRE TUTTE LE RIGHE I CUI TOTAL_BILL SONO > 40
```

```python
mask = tips["total_bill"] > 40
```

```python
tips[mask]
```

```python
# VOGLIAMO ESTRARRE TUTTE LE RIGHE IL CUI SESSO SIA PARI A M
```

```python
mask = tips["sex"] == "Male"
tips[mask]
```

```python
# VOGLIAMO ESTRARRE LE RIGHE IN CORRISPONDENZA DELLE QUALI total_bill > 30 E sex=="Male"
```

```python
mask = (tips["total_bill"]>30) & (tips["sex"]=="Male")
tips[mask]
```

```python
# Possiamo anche vedere un esempio in cui si utilizza l'operatore or |. Questo significa che,
# date due condizioni, otterremo true se è vera solo la prima, solo la seconda, o entrambe.
# Quindi, supponiamo di voler selezionare tutte le righe in corrispondenza delle quali il sesso sia
# F OPPURE smoker sia "Yes"
```

```python
mask = (tips["sex"] == "Female") | (tips["smoker"] == "Yes")
```

```python
tips[mask]
```

```python
# POSSIAMO ANCHE LAVORARE SU UNA SOLA COLONNA: VEDIAMO QUI UN ESEMPIO. 
# CONSIDERIAMO LA VARIABILE time. PER PRIMA COSA VEDIAMO QUALI SONO LE OCCORRENZE DI QUESTA VARIABILE
# CATEGORICA
```

```python
tips['time'].value_counts()
```

```python
# SONO SOLO DINNER O LOUNCH. L'ESEMPIO NON VERREBBE BENE
```

```python
# PROVIAMO CON LA VARIABILE day:
```

```python
tips['day'].value_counts()
```

```python
# VOGLIAMO ESTRARRE LE RIHJE PER LE QUALI LA COLONNA day ASSUME I VALORI Sun oppure Fri
```

```python
mask = (tips["day"]=="Sun") | (tips["day"]=="Fri") 
tips[mask]
```

```python
# VEDIAMO SE è VERO CHE RIMANGONO SOLO LE OCCORRENZE Sun E Fri:
```

```python
solo_sun_fri = tips[mask]
solo_sun_fri["day"].value_counts()
```

```python
# CVD
```

```python
# UN ALTRO MODO PER UTILIZZARE OR | QUANDO SI LAVORA SU CONDIZIONI APPARTENENTI ALLA STESSA VARIABILE
# E' L'OPERATORE isin. RIVEDIAMO L'ULTIMO ESEMPIO 
```

```python
tips
```

```python
mask = tips["day"].isin(["Sun","Fri"])
tips[mask]
```

```python
solo_sun_fri = tips[mask]
solo_sun_fri["day"].value_counts()
```

```python
# RIMOZIONE COLONNA
```

```python
# Supponiamo di voler eliminare la colonna "time". # Per farlo, si utilizza la funzione drop
```

```python
tips.drop("time", axis=1)
```

```python
# Cosa significa axis=1? Semplice, significa che la funzione applicata al df ha a che fare con le colonne.
# Perch√©? Immaginiamo un asse cartesiano. Con x=0 si intende l'asse delle ascisse, e quindi si scorre lungo le colonne,
# con x=1 si intende l'asse delle ordinate, e quindi si scorre lungo le righe. Se vogliamo eliminare una colonna, dovremmo
# eliminare tutti i volori della colonna stessa, e quindi dobbiamo scorrere lungo l'asse delle y, ossia lungo le righe.
# Per questo motivo, si pone x=1. Pi√π avanti vederemo molti esempi per i quali bisogna scegliere se applicare una funzione
# alle colonne (x=1), oppure alle righe (x=0). Per riepilogare: Se vogliamo scorrere lungo le righe utilizzeremo axis=1, se volessimo
# scorrere lungo le colonne avremo axis=0. Se scorriamo lungo le righe applicheremo una funzione alle colonne.
# Se volessimo scurrere lungo le colonne applicheremo una funzione alle righe
```

```python
# Abbiamo quindi eliminato la colonna time!! Ristampiamo il nostro Df
```

```python
tips.head()
```

```python
# Non abbiamo eliminato un tubo. La colonna time √® ancora presente nel Df. Come facciamo ad eliminarla
# definitivamente? Si utilizza un parametro chiamato inplace: tale parametro si inserisce nella nostra funzione drop.
```

```python
tips.drop("time", axis=1, inplace=True)
```

```python
tips.head()
```

```python
# Lo scatter plot è un grafico che ci permette di rappresentare due variabili continue. Esso studia la relazione tra queste due variabili. 
# Una variabile continua è una variabile numerica  che, dati due valori a e b, può assumere infiniti valori in questo intervallo.
# Esempi di variabili numeriche continue:
# - Età
# - Altezza
# - Salario
# - Temperatura
# - Prezzi

# Esempio: Supponiamo di avere a che fare con un'azienda che vende una quantità di merce
# pari a x euro.
# E' possibile, tramite scatter plot, visualizzare la relazione tra il valore della
# in euro della quantità di merce venduta e lo stipendio di ciascun lavoratore.

# Possiamo creare lo scatter plot, mettendo lo stipendio sull'asse delle x e la
# quantità di merce venduta sull'asse delle y. Ogni punto dello scatter plot
# rappresenta un lavoratore.

# VEDIAMO QUINDI FINALMENTE COME SI CREANO GLI SCATTER PLOT TRAMITE LA LIBRERIA 
# SEABORN
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Consideriamo un file dm_office_sales.csv 
```

```python
df = pd.read_csv("dm_office_sales.csv")
df.head()
```

```python
# Vogliamo sapere la relazione tra salary e sales tramite
# scatter plot
```

```python
sns.scatterplot(x="salary", y="sales",data=df)
plt.show()
```

```python
# Possiamo creare una legenda prendendo in considerazione una variabile
# categorica del nostro df. Prendiamo, ad esempio, la division, che indica
# il ramo in cui il lavoratore lavora all'interno dell'azienda. Otterremo il nostro grafico modificato
# nel seguente modo:
```

```python
sns.scatterplot(x="salary", y="sales",data=df,hue="division")
plt.show()
```

```python
# Abbiamo così 3 dimensioni invece di 2. Una dim è il salario, l'altra è la quantità venduta,
# la terza è il settore dell'azienda nel quale il lavoratore lavora.
```

```python
# L'argomento del parametro hue può anche essere di tipo continuo, ma solitamnete non si usa.
```

```python
# Supponiamo di voler cambiare il colore delle categorie. I colori utilizzati finora sono:
# azzurro, arancione, verde, rosso e viola.
# Se volessimo cambiare i colori, dovremmo aggiungere il parametro "palette" al nostro scatterplot.
# I vari palette si trovano nel sito https://matplotlib.org/stable/users/explain/colors/colormaps.html
# Poiché abbiamo a che fare con 5 colori, prendiamo in considerazione quei colori che si trovano
# nella sezione Qualitative colormaps.
# Prendiamo come esempio Set1. I colori sono: Rosso, blu, verde, viola,arancione, giallo etc.
# Quindi:
```

```python
sns.scatterplot(x="salary", y="sales",data=df,hue="division",palette = "Set1")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Ripartiamo dal nostro esempio iniziale: in questo modo facciamo anche ripasso
```

```python
np.random.seed(4)
x = np.random.randint(4,10,8)
x
```

```python
y = 5^x
y
```

```python
np.random.seed(4)
a = np.random.randint(4,70,8)
a 
```

```python
b = a**0.5
b
```

```python
# Rappresentiamo sia la curva sia la retta nello stesso grafico:
```

```python
fig= plt.figure()   # Inizializzaione del canvas
axes= fig.add_axes([0,0,1,1]) # Sia l'asse delle x sia l'asse delle y prendono il 100% 
                              #del canvas.


axes.plot(x,y)
axes.plot(a,b)
```

```python
# Ma che schifo! Comunque, a parte lo schifo del frafico, si noti che abbiamo 
# rappresentate sia la retta sia la curva esponenziale (almeno facciamo finta che lo siano)
# all'interno del nostro grafico. Per rendere il grafico più esplicativo, si può aggiungere
# una legenda all'interno del quadrato. Per farlo, si aggiunge un terzo parametro 
# alla funzione axes.plot. Questo parametro è il parametro label:
```

```python
fig= plt.figure()   # Inizializzaione del canvas
axes= fig.add_axes([0,0,1,1]) # Sia l'asse delle x sia l'asse delle y prendono il 100% 
                              #del canvas.


axes.plot(x,y,label="RETTA SCHIFOSA")
axes.plot(a,b,label="ESPONENZIALE SCHIFOSO")
axes.legend()
```

```python
# Possiamo cambiare la posizione della nostra legenda tramite il 
# parametro loc:
```

```python
fig= plt.figure()   # Inizializzaione del canvas
axes= fig.add_axes([0,0,1,1]) # Sia l'asse delle x sia l'asse delle y prendono il 100% 
                              #del canvas.


axes.plot(x,y,label="RETTA SCHIFOSA")
axes.plot(a,b,label="ESPONENZIALE SCHIFOSO")
axes.legend(loc="upper right")
```

```python
fig= plt.figure()   # Inizializzaione del canvas
axes= fig.add_axes([0,0,1,1]) # Sia l'asse delle x sia l'asse delle y prendono il 100% 
                              #del canvas.


axes.plot(x,y,label="RETTA SCHIFOSA")
axes.plot(a,b,label="ESPONENZIALE SCHIFOSO")
axes.legend(loc="lower right")
```

```python
# Bisogna stare attenti a non sovrapporre le funzioni rappresentate con 
# la legenda. Qui sotto un esempio di sovrapposizione
```
